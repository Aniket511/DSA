```python
class Solution:
    def frequencySort(self, nums: list[int]) -> list[int]:

        hashmap = {}
        for number in nums:
            hashmap[number] = hashmap.get(number, 0) + 1
        
        minHeap = []
        for number, count in hashmap.items():
            heapq.heappush(minHeap, (count, -number))
        
        result = []
        while minHeap:
            count, number = heapq.heappop(minHeap)
            for _ in range(count):
                result.append(-number)
        
        return result

solution = Solution()
print(solution.frequencySort(nums = [1,1,2,2,2,3]))
print(solution.frequencySort(nums = [2,3,1,3,2]))
print(solution.frequencySort(nums = [-1,1,-6,4,5,-6,1,4,1]))
```

```python
class Solution:
    def lastStoneWeight(self, stones: List[int]) -> int:
        
        maxHeap = [-s for s in stones]

        heapq.heapify(maxHeap)

        while len(maxHeap) > 1:
            first = -heapq.heappop(maxHeap)
            second = -heapq.heappop(maxHeap)

            if first > second:
                heapq.heappush(maxHeap, -(first - second))
        
        return 0 if not maxHeap else -maxHeap[0]

solution = Solution()
print(solution.lastStoneWeight(stones = [2,7,4,1,8,1]))
print(solution.lastStoneWeight(stones = [1]))
print(solution.lastStoneWeight(stones = [1,30,400,64,2,4,633,425,634,10,100,324,9,99,99,99,524,1,2,524,252,232,53,424,10, 100, 456,10, 500]))
```

```python
class KthLargest:

    def __init__(self, k: int, nums: List[int]):
        self.minHeap = nums
        heapq.heapify(self.minHeap)
        self.k = k
        while len(self.minHeap) > k:
            heapq.heappop(self.minHeap)

    def add(self, val: int) -> int:
        heapq.heappush(self.minHeap, val)
        if len(self.minHeap) > self.k:
            heapq.heappop(self.minHeap)
        return self.minHeap[0]

solution = KthLargest(4, [7, 7, 7, 7, 8, 3])
print(solution.minHeap)
print(solution.add(2))
print(solution.minHeap)
print(solution.add(12))
print(solution.minHeap)
print(solution.add(9))
print(solution.minHeap)
print(solution.add(13))
print(solution.minHeap)
print(solution.add(6))
print(solution.minHeap)
```

```python
class Solution:
    def topKFrequent(self, words: List[str], k: int) -> List[str]:
        
        hashmap = {}
        for word in words:
            hashmap[word] = hashmap.get(word, 0) + 1

        maxHeap = []
        for word, count in hashmap.items():
            heapq.heappush(maxHeap, (-count, word))

        result = []
        while maxHeap:
            _, word = heapq.heappop(maxHeap)
            result.append(word)
            if len(result) == k:
                break
        
        return result

solution = Solution()
print(solution.topKFrequent(words = ["i","love","leetcode","i","love","coding"], k = 2))
print(solution.topKFrequent(words = ["the","day","is","sunny","the","the","the","sunny","is","is"], k = 4))
```

```python
class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        
        minHeap = []

        for x, y in points:
            curr = x**2 + y**2
            heapq.heappush(minHeap, (curr, x, y))

        result = []
        while minHeap:
            _, x, y = heapq.heappop(minHeap)
            result.append((x, y))
            if len(result) == k:
                break
        
        return result

solution = Solution()
print(solution.kClosest(points = [[1,3],[-2,2]], k = 1))
print(solution.kClosest(points = [[3,3],[5,-1],[-2,4]], k = 2))
```

```python
class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        
        hashmap = {}
        for task in tasks:
            hashmap[task] = hashmap.get(task, 0) + 1
        
        maxHeap = []
        for count in hashmap.values():
            heapq.heappush(maxHeap, -count)

        time = 0
        queue = deque()

        while maxHeap or queue:

            time += 1

            if maxHeap:
                count = heapq.heappop(maxHeap) + 1

                if count:
                    queue.append((count, time + n))
            
            if queue and queue[0][1] == time:
                count, _ = queue.popleft()
                heapq.heappush(maxHeap, count)

        return time

solution = Solution()
solution.leastInterval(['A', 'A','A','B','B','B'], 2)
```

```python
class Solution:
    def longestDiverseString(self, a: int, b: int, c: int) -> str:
        
        maxHeap = []
        for count, letter in [(a, 'a'), (b, 'b'), (c, 'c')]:
            if count:
                heapq.heappush(maxHeap, (-count, letter))
        
        result = ''

        while maxHeap:
            count1, char1 = heapq.heappop(maxHeap)
            if len(result) > 1 and result[-1] == result[-2] == char1:
                if not maxHeap:
                    return result
                count2, char2 = heapq.heappop(maxHeap)
                result += char2
                count2 += 1
                if count2:
                    heapq.heappush(maxHeap, (count2, char2))
            
            result += char1
            count1 += 1
            if count1:
                heapq.heappush(maxHeap, (count1, char1))
        
        return result

solution = Solution()
print(solution.longestDiverseString(a = 1, b = 1, c = 7))
print(solution.longestDiverseString(a = 7, b = 1, c = 0))
print(solution.longestDiverseString(a = 11, b = 1, c = 4))
```

```python
class Solution:
    def reorganizeString(self, s: str) -> str:
        
        hashmap = {}
        for letter in s:
            hashmap[letter] = hashmap.get(letter, 0) + 1

        maxHeap = []
        for letter, count in hashmap.items():
            heapq.heappush(maxHeap, (-count, letter))

        result = ''
        prev = None

        while maxHeap or prev:
            
            if prev and not maxHeap:
                return ''

            count, letter = heapq.heappop(maxHeap)
            result += letter
            count += 1

            if prev:
                heapq.heappush(maxHeap, prev)
                prev = None
            
            if count:
                prev = (count, letter)

        return result

solution = Solution()
print(solution.reorganizeString(s = "aab"))
print(solution.reorganizeString(s = "aaab"))
print(solution.reorganizeString(s = "acab"))
```

```python
class Solution:
    def carPooling(self, trips: List[List[int]], capacity: int) -> bool:
        
        trips.sort(key = lambda trip : trip[1])

        minHeap = []
        currPass = 0

        for numPass, start, end in trips:

            while minHeap and minHeap[0][0] <= start:
                currPass -= heapq.heappop(minHeap)[1]

            currPass += numPass
            if currPass > capacity:
                return False

            heapq.heappush(minHeap, (end, numPass))

        return True

solution = Solution()
print(solution.carPooling(trips = [[2,1,5],[3,3,7]], capacity = 4))
print(solution.carPooling(trips = [[2,1,5],[3,3,7]], capacity = 5))
```

```python
class Solution:
    def getOrder(self, tasks: List[List[int]]) -> List[int]:
        
        for i, task in enumerate(tasks):
            task.append(i)
        
        tasks.sort(key = lambda task : task[0])

        result = []
        minHeap = []
        idx = 0
        time = tasks[0][0]

        while minHeap or idx < len(tasks):

            while idx < len(tasks) and time >= tasks[idx][0]:
                heapq.heappush(minHeap, (tasks[idx][1], tasks[idx][2]))
                idx += 1

            if not minHeap:
                time = tasks[idx][0]

            else:
                procTime, indx = heapq.heappop(minHeap)
                time += procTime
                result.append(indx)

        return result

solution = Solution()
print(solution.getOrder(tasks = [[1,2],[2,4],[3,2],[4,1]]))
print(solution.getOrder(tasks = [[7,10],[7,12],[7,5],[7,4],[7,2]]))
```

```python
class Twitter:

    def __init__(self):
        self.tweetTime = 0
        self.tweetMap = defaultdict(list)
        self.followMap = defaultdict(set)

    def postTweet(self, userId: int, tweetId: int) -> None:
        self.tweetMap[userId].append((self.tweetTime, tweetId))
        self.tweetTime -= 1

    def getNewsFeed(self, userId: int) -> List[int]:
        result = []
        minHeap = []

        self.followMap[userId].add(userId)
        for followeeId in self.followMap[userId]:
            if followeeId in self.tweetMap:
                idx = len(self.tweetMap[followeeId]) - 1
                count, tweetId = self.tweetMap[followeeId][idx]
                heapq.heappush(minHeap, (count, tweetId, followeeId, idx - 1))

        while minHeap and len(result) < 10:
            count, tweetId, followeeId, idx = heapq.heappop(minHeap)
            result.append(tweetId)
            if idx >= 0:
                count, tweetId = self.tweetMap[followeeId][idx]
                heapq.heappush(minHeap, (count, tweetId, followeeId, idx - 1))
        
        return result

    def follow(self, followerId: int, followeeId: int) -> None:
        self.followMap[followerId].add(followeeId)

    def unfollow(self, followerId: int, followeeId: int) -> None:
        if followeeId in self.followMap[followerId]:
            self.followMap[followerId].remove(followeeId)
```

```python
class MedianFinder:
    def __init__(self):
        self.small = []
        self.large = []

    def addNum(self, num: int) -> None:
        if self.large and num > self.large[0]:
            heapq.heappush(self.large, num)
        else:
            heapq.heappush(self.small, -1 * num)

        if len(self.small) > len(self.large) + 1:
            val = -1 * heapq.heappop(self.small)
            heapq.heappush(self.large, val)
        if len(self.large) > len(self.small) + 1:
            val = heapq.heappop(self.large)
            heapq.heappush(self.small, -1 * val)

    def findMedian(self) -> float:
        if len(self.small) > len(self.large):
            return -1 * self.small[0]
        elif len(self.large) > len(self.small):
            return self.large[0]
        return (-1 * self.small[0] + self.large[0]) / 2.0
```

```python
class Solution:
    def findMaximizedCapital(self, k: int, w: int, profits: list[int], capital: list[int]) -> int:
        # Min-heap by capital requirement
        available = []  # max-heap by profit (negate for Python)
        projects = sorted(zip(capital, profits))  # sort by capital needed
        
        idx = 0
        for _ in range(k):
            # Push all affordable projects into max-heap
            while idx < len(projects) and projects[idx][0] <= w:
                heapq.heappush(available, -projects[idx][1])
                idx += 1
            
            # No affordable projects left
            if not available:
                break
            
            # Pick most profitable available project
            w += -heapq.heappop(available)
        
        return w
```

```python
class Solution:
    def trapRainWater(self, heightMap):
        if not heightMap or len(heightMap) < 3 or len(heightMap[0]) < 3:
            return 0

        numOfRows = len(heightMap)
        numOfColumns = len(heightMap[0])
        visited = [[False] * numOfColumns for _ in range(numOfRows)]
        minHeap = []

        # Push left and right columns (all rows)
        for row in range(numOfRows):
            for column in (0, numOfColumns - 1):
                heapq.heappush(minHeap, (heightMap[row][column], row, column))
                visited[row][column] = True

        # Push top and bottom rows, skipping corners already added above
        for column in range(1, numOfColumns - 1):
            for row in (0, numOfRows - 1):
                heapq.heappush(minHeap, (heightMap[row][column], row, column))  # row is row, column is column
                visited[row][column] = True

        water = 0

        while minHeap:
            height, x, y = heapq.heappop(minHeap)

            for dx, dy in [(0,1),(0,-1),(1,0),(-1,0)]:
                nx, ny = x + dx, y + dy

                if 0 <= nx < numOfRows and 0 <= ny < numOfColumns and not visited[nx][ny]:
                    visited[nx][ny] = True
                    water += max(0, height - heightMap[nx][ny])
                    heapq.heappush(minHeap, (max(height, heightMap[nx][ny]), nx, ny))

        return water
```
